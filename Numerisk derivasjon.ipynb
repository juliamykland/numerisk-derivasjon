# Numerisk derivasjon
## Funksjon for å derivere


definisjonen av den deriverte:
$$f'(a) = \lim_{\Delta x \to 0}\frac{f(a + \Delta x) - f(a)}{\Delta x}$$
- den deriverte til en funksjon er et mål på endringen i funkjsonsverdien f(x) som følge av en veldig liten endring i variablen x
- momentan endring
- stigningstall til tangenten i punktet

definisjon av numerisk derivasjon: $$f'(a)≈\frac{f(a + \Delta x) - f(a)}{\Delta x}$$
- vi velger en liten $\Delta x$ for å numerisk beregne den deriverte i a. Man kan prøve å gjære $\Delta x$ mindre for å se om den deriverte endres. Hvis den deriverte ikke endres mer når $\Delta x$ blir mindre, så er den numeriske derivasjonen god nok.

# Numerisk

```python
from pylab import *

def f(x):
    return x**2

def fder(f, x, dx=1E-8):
    df = (f(x + dx) - f(x))/dx
    return df

df = fder(f, 1)

print(df)

```

# Derivasjon av diskrete data

```python
from pylab import *
t = [5, 10, 15, 30, 50, 60]
s = [115, 190, 390, 600, 1000, 1200]
v = []
v.append(0)

for i in range(len(s)-1):
    sder = (s[i+1] - s[i])/(t[i+1]-t[i])
    v.append(sder)
    
plot(t,v)

```

# Derivasjon av heisdata

```python
from pylab import *

data = loadtxt('elevator.txt',skiprows=1, delimiter=',')

#print(data)

tid = data[:,0]
hoyde = data[:,2]

"""
plot(tid, hoyde)
xlabel('tid (s)')
ylabel('høyde (m)')
"""

v = []
v.append(0)

for i in range(0,47):
    fart = (hoyde[i+1]-hoyde[i])/(tid[i+1]-tid[i])
    v.append(fart)
   

print("Fart: ", fart, "m/s")


plot(tid,v)
title('Fartsgraf')
xlabel('tid (s)')
ylabel('fart (m/s)')
show()

a = []
a.append(0)

for i in range(0,47):
    akselerasjon = (v[i+1]-v[i])/(tid[i+1]-tid[i])
    a.append(akselerasjon)
    
plot(tid,a)
title('Akselerasjonsgraf')
xlabel('tid (s)')
ylabel('akselerasjon (m/s^2)')
show()
```
