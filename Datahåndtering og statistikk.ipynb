## Sunspots
### lesing og plotting av data

Vi kan lese filer vha. funksjonen _loadtxt_:
data = loadtxt('filnavn',dtype='...', skiprows=... , delimiter='...', usecols=(...))

Gjennomsnitt: summen av en tall rekke delt på antall elementer i rekka.


Standaravvik: et mål for hvor mye elementene i tallrekka aviker fra gjennomsnittet. 

```python
from pylab import *

data = loadtxt('sunspots.txt',skiprows=1, delimiter=',')

print(data)

year = data[:,0]
sunactivity = data[:,1]

print(year)
print(sunactivity)

plot(year, sunactivity)
xlabel('year')
ylabel('sunactivity')
show()


snitt = mean(sunactivity)
avvik = std(sunactivity)

print("Gjenomsnitt: ",snitt, "Standaravvik: ",avvik)
```
